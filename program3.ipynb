Approach: Our first approach was the Longest Common Substring. We then used iterations to compare characters from both words and track whenever a matching substring appeared. Our approach was the Longest Unique Substring, which we primarily used to compare efficiency and determine which method worked better for this problem.

```python
# Naive recursive approach
def leastCommon(current_result, s1, s2, m, n):
    if m == 0 or n == 0 or s1[m - 1] != s2[n - 1]:
        return current_result
    new_result =  s1[m - 1] + current_result
    return leastCommon(new_result, s1, s2, m - 1, n - 1)  

def maximumSubstring(s1, s2):
    res = ""
    m = len(s1)
    n = len(s2)

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            out = leastCommon("", s1, s2, i, j)
            if len(out) > len(res):
                res = out
    return res
	
# Memoized approach	
def leastCommonMemoized(s1, s2, m, n, memo):
    if m == 0 or n == 0:
        return 0

    if memo[m][n] is not None:
        return memo[m][n]

    if s1[m - 1] == s2[n - 1]:
        memo[m][n] = 1 + leastCommonMemoized(s1, s2, m - 1, n - 1, memo)
    else:
        memo[m][n] = 0
    return memo[m][n]

def maximumSubstringMemoized(s1, s2):
    m = len(s1)
    n = len(s2)
    memo = [[None] * (n + 1) for _ in range(m + 1)]

    max_len = 0
    end = 0

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            length = leastCommonMemoized(s1, s2, i, j, memo)
            if length > max_len:
                max_len = length
                end = i
    return s1[end - max_len:end]
	

# Dynamic Programming approach
def maximumSubstringDP(s1, s2):
    m, n = len(s1), len(s2)
    
    dp = [[0 for _ in range(n + 1)] for _ in range(m + 1)]
    
    max_len = 0
    end_pos = 0  

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
                if dp[i][j] > max_len:
                    max_len = dp[i][j]
                    end_pos = i
            else:
                dp[i][j] = 0 

    return s1[end_pos - max_len:end_pos]
```

```python
#Test cases 
list = ['orange', 'banana', 'apple', 'pineapple']

print('Naive Recursive Approach:')
print(maximumSubstring(list[0], list[1]))
print(maximumSubstring(list[1], list[2]))
print(maximumSubstring(list[2], list[3]))
print(maximumSubstring(list[0], list[3]))

print('Memoized Approach:')
print(maximumSubstringMemoized(list[0], list[1]))
print(maximumSubstringMemoized(list[1], list[2]))
print(maximumSubstringMemoized(list[2], list[3]))
print(maximumSubstringMemoized(list[0], list[3]), "\n")

print('Dynamic Programming Approach:')
print(maximumSubstringDP(list[0], list[1]))
print(maximumSubstringDP(list[1], list[2]))
print(maximumSubstringDP(list[2], list[3]))
print(maximumSubstringDP(list[0], list[3]), "\n")
```

## Justification of Correctness

(proof by induction)

### Algorithm 1: Backtracking Approach

leastCommon helper:

Base case (i=0/j=0): 

One string is empty, no suffix exists. Returns "". (Correct)

Inductive Hypothesis: Assume for all smaller indicies (i-1, j-1), the function correctly computes the longest common suffix. 

Inductive Step: 

Two cases - Case 1 - Match: (s1[i-1] == s2[j-1]) - Extend suffix: (return s1[i-1] + leastCommon(i-1, j-1)). By the hypothesis, the recursive call is correct, so the result is correct. Case 2 - No match: (return ""). Correct because substrings must be contiguous. 

Therefore, leastCommon helper is correct

Main algorithm:

We iterate over all (i, j) pairs. 

After checking first k pairs, res stores the longest substring among them. For the next pair (i, j), compoute the longest substring ending there and update res = max(res, candidate). Every substring must end at some (i, j), and we check all of them.

By induction, this algorithm returns the global longest common substring. 

### Algorithm 2: Memorized Approach

Base case:

When a subproblem (i, j) is computed for the first time, if i == 0 or j == 0, the result is "". This is correct because an empty string has no common substring. 

Inductive Hypothesis: Assume that all previously computed subproblems stored in the memo table are correct.

Inductive Step: 

When computing a new subproblem (i, j), if the value is already in the memo table, it is correct. Otherwise, the algorithm computes it using the same recurrence as the first algorithm (already proved correct). Then it stores the correct result in the memo table. 

By induction, all memorized values are correct, so the final answer is correct

### Algorithm 3: Dynamic Programming Approach

Let dp[i][j] = length of longest common substring ending at s1[i-1] and s2[j-1]

Base case:

If i == or j == 0, then dp[i][j] = 0, which is correct because an empty string cannot have a common substring

Inductive Hypothesis: Assume all values dp[x][y] for smaller indicies are correct. 

Inductive Step:

For dp[i][j], we consider

Case 1: Characters match (s1[i-1] == s2[j-1]). Then, dp[i][j] = dp[i-1][j-1] + 1 and by the hypothesis, dp[i-1][j-1] is correct, so extending it by one is correct.

Case 2: Characters do not match (dp[i][j] = 0). This is correct because substrings must be contiguous, so the match resets. 

Final step: The algorithm returns max(dp[i][j]). This works because every substring must end at some (i, j)

Therefore, by induction, all DP values are correct, so the algorithm correctly returns the largest common substring.

## Asymptotic Runtime

### Algorithm 1: Backtracking Approach

Step 1: Outer loops

The algorithm checks all pairs (i,j) - (i = 1...n, j = 1...m), so total iterations is n*m

Step 2: Cost of leastCommon(i, j)

In the worst case, characters match repeatedly ('aaaaaa...'). The recursion goes back up to min(i, j). Worst case: O(min(n, m))

For each of the n*m pairs, we do up to O(min(n, m)) work. Therefore, 

#### T(n) = O(n * m * min(n,m))

(if n = m, this becomes O(n^3))

### Algorithm 2: Memorized Approach

Step 1: Number of subproblems

Possible states: (i,j) -> n*m

Step 2: Work per subproblem

Each computation has one comparison, one lookup, and one addition (O(1))

Total runtime: O(n * m)

### Algorithm 3: Dynamic Programming Approach

Step 1: DP table size

We build a table dp[n+1][m+1]. This has (n+1)(m+1) total entries which is O(nm)

Step 2: Work per call

Each cell does one comparison and one assignment (constant time)

Total Runtime: O(nm)
