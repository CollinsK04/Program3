Approach: Our first approach was the Longest Common Substring. We then used iterations to compare characters from both words and track whenever a matching substring appeared. Our approach was the Longest Unique Substring, which we primarily used to compare efficiency and determine which method worked better for this problem.

```python
# Naive recursive approach
def leastCommon(current_result, s1, s2, m, n):
    if m == 0 or n == 0 or s1[m - 1] != s2[n - 1]:
        return current_result
    new_result =  s1[m - 1] + current_result
    return leastCommon(new_result, s1, s2, m - 1, n - 1)  

def maximumSubstring(s1, s2):
    res = ""
    m = len(s1)
    n = len(s2)

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            out = leastCommon("", s1, s2, i, j)
            if len(out) > len(res):
                res = out
    return res
	
# Memoized approach	
def leastCommonMemoized(s1, s2, m, n, memo):
    if m == 0 or n == 0:
        return 0

    if memo[m][n] is not None:
        return memo[m][n]

    if s1[m - 1] == s2[n - 1]:
        memo[m][n] = 1 + leastCommonMemoized(s1, s2, m - 1, n - 1, memo)
    else:
        memo[m][n] = 0
    return memo[m][n]

def maximumSubstringMemoized(s1, s2):
    m = len(s1)
    n = len(s2)
    memo = [[None] * (n + 1) for _ in range(m + 1)]

    max_len = 0
    end = 0

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            length = leastCommonMemoized(s1, s2, i, j, memo)
            if length > max_len:
                max_len = length
                end = i
    return s1[end - max_len:end]
	

# Dynamic Programming approach
def maximumSubstringDP(s1, s2):
    m, n = len(s1), len(s2)
    
    dp = [[0 for _ in range(n + 1)] for _ in range(m + 1)]
    
    max_len = 0
    end_pos = 0  

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
                if dp[i][j] > max_len:
                    max_len = dp[i][j]
                    end_pos = i
            else:
                dp[i][j] = 0 

    return s1[end_pos - max_len:end_pos]
```

```python
#Test cases 
list = ['orange', 'banana', 'apple', 'pineapple']

print('Naive Recursive Approach:')
print(maximumSubstring(list[0], list[1]))
print(maximumSubstring(list[1], list[2]))
print(maximumSubstring(list[2], list[3]))
print(maximumSubstring(list[0], list[3]))

print('Memoized Approach:')
print(maximumSubstringMemoized(list[0], list[1]))
print(maximumSubstringMemoized(list[1], list[2]))
print(maximumSubstringMemoized(list[2], list[3]))
print(maximumSubstringMemoized(list[0], list[3]), "\n")

print('Dynamic Programming Approach:')
print(maximumSubstringDP(list[0], list[1]))
print(maximumSubstringDP(list[1], list[2]))
print(maximumSubstringDP(list[2], list[3]))
print(maximumSubstringDP(list[0], list[3]), "\n")
```
